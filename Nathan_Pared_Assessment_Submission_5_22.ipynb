```python
#making sure that the data are the data types that I need for my analysis.
clean_star_data = raw_star_data.replace('?', 0)
clean_star_data = clean_star_data.astype({'HoursPerWeek':'int64', 'TotalHours':'int64'})
clean_star_data
```

### Getting roughly %25 of the data to analyze

```python
samp_data_1 = clean_star_data.head(850)
samp_data_1
```

```python
#we are going to ignore the age of a player for the main analysis
samp_data_1a = samp_data_1.drop(columns=['Age'])
samp_data_1a
```

### Determining what makes a high ranked player based off of LeagueIndex

```python
# high ranked league index
high_ind = samp_data_1a[samp_data_1a['LeagueIndex'] >= 5]
high_ind
```

```python
high_ind.shape
```

```python
#low ranked league index
low_ind = samp_data_1a[samp_data_1a['LeagueIndex'] <= 4]
low_ind
```

```python
low_ind.shape
```

```python
high_ind.describe(include='all')
```

```python
low_ind.describe(include='all')
```

#### Information based of the described info of the higher indexed players are as follows <i>(based on the average of all the columns)</i>:
1. Higher APM
2. Way higher hotkey usage, selection, and uniqueness
3. Way higher Minimap Attacks, minimap right clicks
4. Higher attention shifts (PACs)
5. Smaller gaps between PAcs
6. Samll difference in attacks within each PAC
7. Map is more explored
8. More workers are made
9. Slightly more unique units, complex units, and complex abilities used
10. The total hours for higher indexed players are twice that than the lower indexed players

<b>With this information gathered, we can say that the higher indexed players would have better stats in every column than the lower indexed players. Therefore, higher ranked players would have better stats in every category of the game.</b>

### Graphs of the differences for main columns

Graph of the total hours:

```python
high_avg_big = [high_ind['TotalHours'].mean(), high_ind['APM'].mean(),high_ind['TotalMapExplored'].mean()]
high_avg_small = [high_ind['SelectByHotkeys'].mean(), high_ind['MinimapRightClicks'].mean(), high_ind['NumberOfPACs'].mean(), high_ind['WorkersMade'].mean()]
```

```python
low_avg_big = [low_ind['TotalHours'].mean(), low_ind['APM'].mean(), low_ind['TotalMapExplored'].mean()]
low_avg_small = [low_ind['SelectByHotkeys'].mean(), low_ind['MinimapRightClicks'].mean(), low_ind['NumberOfPACs'].mean(), low_ind['WorkersMade'].mean()]
```

```python
label_stats_big = ['TotalHours', 'APM', 'TotalMapExplored']
plt.bar(label_stats_big, high_avg_big, label='high_index')
plt.bar(label_stats_big, low_avg_big, bottom=high_avg_big, label='low_index')
plt.xticks(rotation=90)
plt.xlabel('Categories')
plt.ylabel('Averages')
plt.title('Differences in Category Stats')
plt.legend()
plt.show()
```

```python
label_stats_small = ['SelectByHotkeys', 'MinimapRightClicks', 'NumberOfPACs', 'WorkersMade']
plt.bar(label_stats_small, high_avg_small, label='high_stats')
plt.bar(label_stats_small, low_avg_small, bottom=high_avg_small, label='low_stats')
plt.xticks(rotation=90)
plt.xlabel('Categories')
plt.ylabel('Averages')
plt.title('Differences in Category Stats')
plt.legend()
plt.show()
```
